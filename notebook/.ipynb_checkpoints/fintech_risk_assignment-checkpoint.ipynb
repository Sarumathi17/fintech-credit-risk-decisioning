# FINANCIAL RISK DECISIONING FRAMEWORK

## OBJECTIVE
The objective of this project is to build a credit risk decisioning framework for SMEs.
Given a company’s financial ratios, we aim to:

1. Estimate the probability of financial distress (PD)
2. Convert the PD into interpretable risk tiers (A–D)
3. Design an Expected Credit Loss (ECL)–based approve/decline policy

This project is framed as a business decisioning problem rather than a pure
classification task.

## DOMAIN ANALYSIS – SME Credit Risk

In fintech lending, credit risk refers to the likelihood that a borrower will
fail to meet repayment obligations.

For SME credit decisioning, risk is typically evaluated using:
- **Probability of Default (PD):** Likelihood of financial distress
- **Loss Given Default (LGD):** Portion of exposure not recovered after default
- **Exposure at Default (EAD):** Amount at risk when default occurs

These components are combined to estimate **Expected Credit Loss (ECL)**:

ECL = PD × LGD × EAD

In this project:
- Bankruptcy is used as a proxy for default
- Financial ratios serve as indicators of liquidity, profitability, leverage, and efficiency
- Model outputs are translated into risk tiers and an approve/decline policy to support real-world lending decisions

### 6. MODEL BUILDING

### 6.1 Baseline Model – Logistic Regression

Logistic Regression is used as a baseline probability of default (PD) model
due to its interpretability and widespread use in credit risk modeling.


```python
#Feature scaling

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lr = LogisticRegression(
    max_iter=1000,
    random_state=42
)

lr.fit(X_train_scaled, y_train)
```

```python
#Predict PD on Test Set
y_test_pred_proba_lr = lr.predict_proba(X_test_scaled)[:, 1]
```

```python
#Model Evaluation – ROC-AUC
roc_auc_lr = roc_auc_score(y_test, y_test_pred_proba_lr)
roc_auc_lr
```

```python
#PR-AUC
precision, recall, _ = precision_recall_curve(
    y_test, y_test_pred_proba_lr
)

pr_auc_lr = auc(recall, precision)
pr_auc_lr
```

### Baseline Model Performance Interpretation

The Logistic Regression model achieves a ROC-AUC of 0.866,
indicating strong ranking ability between distressed and healthy firms.

The PR-AUC of 0.233 significantly exceeds the base bankruptcy rate
(~3%), demonstrating meaningful predictive power for the minority class.

This establishes a strong and interpretable benchmark for comparison
with more complex models.


### 6.2 Improved Model – XGBoost

Given the non-linear risk patterns observed during EDA,
a tree-based gradient boosting model is implemented to capture
complex feature interactions.

This model aims to improve minority class detection
while maintaining strong ranking performance.

```python
from xgboost import XGBClassifier
```

#### Model Configuration

The model is regularized to prevent overfitting by:
- Limiting tree depth
- Using subsampling
- Applying L2 regularization


```python
xgb = XGBClassifier(
    n_estimators=150,
    max_depth=3,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_lambda=5,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
```

#### Model Training

The model is trained only on the training dataset to avoid data leakage.


```python
xgb.fit(X_train, y_train)
```

#### Predict Probability of Distress (PD)

The predicted probability for class 1 represents the
Probability of Default (PD).


```python
y_train_pred_xgb = xgb.predict_proba(X_train)[:, 1]
y_test_pred_xgb = xgb.predict_proba(X_test)[:, 1]
```

#### Model Performance Evaluation

Performance is evaluated using:

- ROC-AUC (ranking ability)
- PR-AUC (minority class detection)


```python
from sklearn.metrics import roc_auc_score

train_roc_xgb = roc_auc_score(y_train, y_train_pred_xgb)
test_roc_xgb = roc_auc_score(y_test, y_test_pred_xgb)

train_roc_xgb, test_roc_xgb
```

```python
from sklearn.metrics import precision_recall_curve, auc

precision_xgb, recall_xgb, _ = precision_recall_curve(
    y_test, y_test_pred_xgb
)

pr_auc_xgb = auc(recall_xgb, precision_xgb)

pr_auc_xgb
```

#### Overfitting Check

Training performance is compared with test performance.
A moderate gap indicates acceptable generalization.


### XGBoost Model Performance Interpretation

The XGBoost model significantly improves both ROC-AUC
and PR-AUC compared to Logistic Regression.

The large increase in PR-AUC indicates substantially
improved detection of distressed firms under class imbalance,
making XGBoost the preferred final PD model.


#### Cross-Validation Stability Check

5-fold stratified cross-validation is performed to ensure
that performance is stable across different data splits.


```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    xgb,
    X_train,
    y_train,
    cv=5,
    scoring='roc_auc'
)

cv_scores.mean()
```

#### Cross-Validation Interpretation

The 5-fold stratified cross-validation yields a mean ROC-AUC of approximately 0.937.

This indicates that the model maintains strong ranking performance
across different training-validation splits and is not dependent on
a single train-test partition.

Although the training ROC is higher, the cross-validation results
suggest that the model generalizes well and does not exhibit
severe overfitting.


### Final Model Selection

Given the substantial improvement in both ROC-AUC and PR-AUC,
along with stable cross-validation performance, XGBoost is selected
as the final Probability of Default (PD) model for subsequent
risk tiering and decision policy design.


### 7. COMPANY RISK PROFILE OUTPUT

To translate model predictions into actionable business outputs,
a structured risk profile is created for each company, including:

- pd: Predicted Probability of Default
- risk_tier: A/B/C/D classification
- top_reasons: Top 3 drivers influencing predicted risk

### 7.1 Tier Design Methodology

Risk tiers are defined using percentiles of the predicted PD distribution
on the test set.

This approach ensures:

- Balanced segmentation across the portfolio
- Stability under different score distributions
- Relative risk ranking consistency

```python
#Create PD DataFrame
risk_df = X_test.copy()

risk_df["PD"] = y_test_pred_xgb
risk_df["Actual"] = y_test.values

risk_df.head()
```

```python
#Define Tier Cutoffs
p25 = np.percentile(risk_df["PD"], 25)
p50 = np.percentile(risk_df["PD"], 50)
p75 = np.percentile(risk_df["PD"], 75)

p25, p50, p75
```

### 7.2 Risk Tier Assignment

```python
#Assign Risk Tiers
def assign_tier(pd):
    if pd <= p25:
        return "A"
    elif pd <= p50:
        return "B"
    elif pd <= p75:
        return "C"
    else:
        return "D"

risk_df["Risk_Tier"] = risk_df["PD"].apply(assign_tier)

risk_df[["PD", "Risk_Tier"]].head()
```

```python
risk_df.groupby("Risk_Tier")["Actual"].mean()
```

### 7.3 Model Explainability (Top Reasons)

To provide interpretability at the company level, SHAP (SHapley Additive exPlanations)
is used to identify the top contributing features driving each firm's
predicted Probability of Default (PD).

These reason codes enhance transparency and support credit decision-making.

```python
import shap

explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X_test)
```

```python
shap_df = pd.DataFrame(
    shap_values,
    columns=X_test.columns,
    index=X_test.index
)
```

```python
def get_top_3_reasons(row):
    # Sort features by absolute SHAP value
    top_features = row.abs().sort_values(ascending=False).head(3).index.tolist()
    return ", ".join(top_features)

risk_df["Top_Reasons"] = shap_df.apply(get_top_3_reasons, axis=1)

pd.set_option('display.max_colwidth', None)
risk_df[["PD", "Risk_Tier", "Top_Reasons"]].head()
```

### Tier Performance Interpretation

The observed bankruptcy rate increases materially from Tier A to Tier D,
demonstrating clear risk separation across segments.

- Tiers A and B exhibit near-zero observed distress, indicating strong financial stability.
- Tier C shows moderate risk concentration.
- Tier D exhibits a significantly elevated bankruptcy rate (\~12.7%),
  compared to the overall portfolio default rate (~3%).

This monotonic increase confirms that the model effectively concentrates
credit risk within the highest tier.

Such separation validates the tiering framework as a reliable tool
for portfolio segmentation and credit decisioning.

### 8. EXPECTED CREDIT LOSS (ECL) & DECISION POLICY

### 8.1 Business Assumptions

To translate predicted probabilities into financial impact,
the following simplified credit risk assumptions are made:

- **EAD (Exposure at Default)** = USD 10,000
- **LGD (Loss Given Default)** = 60%

These assumptions allow conversion of model PD outputs
into Expected Credit Loss (ECL).


```python
#Define Business Assumptions
EAD = 10000
LGD = 0.60
```

### 8.2 Expected Credit Loss Calculation

Expected Credit Loss (ECL) is computed as:

ECL = PD × LGD × EAD

This represents the expected financial loss per approved company.

```python
#Compute ECL Per Company
risk_df["ECL"] = risk_df["PD"] * LGD * EAD

risk_df[["PD", "Risk_Tier", "ECL"]].head()
```

```python
risk_df["ECL"].describe()
```

### 8.3 Decision Policy Design

A simple approval rule is defined:

Approve a company if its ECL is below a predefined threshold.

This approach links model output directly to financial risk exposure.


```python
ECL_threshold = 100

risk_df["Decision"] = np.where(
    risk_df["ECL"] < ECL_threshold,
    "Approve",
    "Decline"
)
```

#### Threshold Selection Rationale

An ECL threshold of USD 100 was selected as the approval cutoff.

This value represents the maximum acceptable expected loss per company
under the simplified portfolio assumptions (LGD = 60%, EAD = USD 10,000).

The threshold was chosen to balance:

- **Capital protection** — limiting exposure to high-risk firms
- **Portfolio growth** — maintaining a strong approval rate

```python
approval_rate = (risk_df["Decision"] == "Approve").mean()

total_expected_loss = risk_df.loc[
    risk_df["Decision"] == "Approve", "ECL"
].sum()

expected_defaults = risk_df.loc[
    risk_df["Decision"] == "Approve", "PD"
].sum()

approval_rate, total_expected_loss, expected_defaults
```

### 8.4 Portfolio-Level Results

Applying the ECL-based policy on the test set yields:

- Approval Rate: ~82.7%
- Total Expected Loss (approved firms): ~USD 25.8k
- Estimated Expected Defaults: ~4.3 companies

These metrics quantify the trade-off between portfolio growth
and capital protection.

### 8.5 Confusion Matrix Analysis

```python
pd_threshold = ECL_threshold / (LGD * EAD)

risk_df["Predicted_Class"] = np.where(
    risk_df["PD"] >= pd_threshold,
    1,  # Predict bankrupt
    0   # Predict healthy
)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    risk_df["Actual"],
    risk_df["Predicted_Class"]
)

cm
```

```python
import pandas as pd

cm_df = pd.DataFrame(
    cm,
    index=["Actual_Healthy", "Actual_Bankrupt"],
    columns=["Pred_Healthy", "Pred_Bankrupt"]
)

cm_df
```

#### Confusion Matrix Interpretation

The ECL-based threshold results in:

- 1406 healthy firms correctly approved.
- 51 distressed firms correctly declined.
- Only 4 distressed firms incorrectly approved (False Negatives).

This corresponds to a bankrupt recall rate of ~92.7%,
indicating strong risk detection while maintaining an
approval rate above 80%.

The decision rule effectively limits exposure to high-risk
firms while preserving portfolio growth.


### 9. PROBABILITY CALIBRATION

### 9.1 Why Calibration Matters

While ROC-AUC evaluates ranking ability, credit risk models must also
produce well-calibrated probabilities.

Calibration ensures that predicted PD values align with observed default
frequencies in the portfolio.

For example, if a group of firms has predicted PD ≈ 10%,
approximately 10% of them should actually default.

Well-calibrated PD estimates are critical because they directly
impact Expected Credit Loss (ECL) calculations.


### 9.2 Calibration Curve

The calibration curve compares:

- Mean predicted PD (x-axis)
- Observed default rate (y-axis)

The diagonal line represents perfect calibration.

```python
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

prob_true, prob_pred = calibration_curve(
    y_test,
    y_test_pred_xgb,
    n_bins=10
)

plt.figure()
plt.plot(prob_pred, prob_true, marker='o')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("Mean Predicted PD")
plt.ylabel("Observed Default Rate")
plt.title("Calibration Curve – XGBoost")
plt.show()
```

### 9.3 Brier Score Evaluation

The Brier score measures the mean squared difference between
predicted probabilities and actual outcomes.

Lower values indicate better probability accuracy.

```python
from sklearn.metrics import brier_score_loss
brier_score_loss(y_test, y_test_pred_xgb)
```

### 9.4 Interpretation

The calibration curve closely follows the diagonal line,
indicating that predicted PD values are reasonably aligned
with observed default rates.

Minor deviations at higher PD bins are expected due to
smaller sample sizes in extreme risk segments.

The Brier score (\~0.022) confirms good probability accuracy
relative to the base default rate (~3%).

Overall, the model demonstrates acceptable calibration,
supporting its use for ECL-based decisioning.
